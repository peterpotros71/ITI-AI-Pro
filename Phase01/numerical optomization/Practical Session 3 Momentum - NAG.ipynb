#### Use your final version of the implemented Batch GD algorithm for single variable to find the best line filt of the given data.
#### Coose the suitable number of iterations, learning rate, and stop criteria.
#### Calculate r2 score. Shouldn't below 0.9
#### Plot the required curves (loss-epochs, loss-theta0, loss-theta1, all fitted lines per epoch (single graph) , best fit line)

```python
def batch_gd(inputs, targets, learn_rate, epoch):
    losses = []
    thetas = []
    theta = np.atleast_2d(np.zeros((inputs.shape[1]))).T
    for i in range(epoch):
        print("epoch: " + str(i+1))
        outputs = np.dot(inputs, theta)
        loss = np.mean((outputs - targets)**2) / 2
        print("loss: " + str(loss))
        losses.append(loss)
        grads = np.atleast_2d(inputs).T.dot(outputs-targets) / len(targets)
        theta = theta - learn_rate * grads
        thetas.append(theta)
        #thetas1.append(theta[1])
        jac = np.linalg.norm(grads, 2)   
        print("jac: " + str(jac))
        if jac < .0001 and jac > -.0001:
            break        
    return thetas , losses 
```

```python
thetas, losses = batch_gd(inputs, y, .01, 10000)
```

### Update your Batch GD for one variable  implementation to be NAG and check your results

```python
def nag_grad(inputs, targets, learn_rate, epoch, gamma):
    theta = np.atleast_2d(np.zeros((inputs.shape[1]))).T
    losses = []
    thetas = []
    v_history = 0
    for i in range(epoch):
        print("epoch: " + str(i+1))
        theta_temp = theta - gamma * v_history
        output = np.dot(inputs, theta_temp)
        loss = np.mean((output - targets)**2) / 2
        print("loss: " + str(loss))
        losses.append(loss)
        temp_grads = np.atleast_2d(inputs).T.dot(output-targets) / len(targets)
        theta = theta_temp - learn_rate * temp_grads
        thetas.append(theta)
        v_history = learn_rate * temp_grads + gamma * v_history
        jac = np.linalg.norm(temp_grads, 2)   
        print("jac: " + str(jac))    
        if jac < .0001 and jac > -.0001:
            break 
    return thetas, losses        
    #plt.ylim(-1,5)
    #plt.plot(losses);
```
