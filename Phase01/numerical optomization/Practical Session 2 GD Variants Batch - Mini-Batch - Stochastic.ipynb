```python
x = np.linspace(0,20, 50).reshape(-1,1)
```

```python
y = -2 * x + 1
```

```python
inputs = np.atleast_2d(np.ones(50)).T
inputs = np.append(inputs, x, axis=1)
```

#### Use your code from practical session 1 and create a function to perform GD for single variable linear regression.
#### The function should take the following input parameters:
##### Input data (X), Target Labels (y), Learning Rate (alpha), Maximum number of iterations
#### The function should return the following:
##### All Theta_0 and Theta_1 values till converge, and equivellant Losses, all hypothesis output, and optimum values of Theta_0 and Theta_1
#### Note: you can choose any suitable stop criteria.

```python
def batch_gd(inputs, targets, learn_rate, epoch):
    losses = []
    thetas = []
    theta = np.atleast_2d(np.zeros((inputs.shape[1]))).T
    #np.hstack(inputs, targets)
    for i in range(epoch):
        print("epoch: " + str(i+1))
        outputs = np.dot(inputs, theta)
        loss = np.mean((outputs - targets)**2) / 2
        print("loss: " + str(loss))
        losses.append(loss)
        grads = np.atleast_2d(inputs).T.dot(outputs-targets) / len(targets)
        theta = theta - learn_rate * grads
        thetas.append(theta)
        #thetas1.append(theta[1])
        jac = np.linalg.norm(grads, 2)   
        print("jac: " + str(jac))
        if jac < .0001 and jac > -.0001:
            break        
    return thetas , losses       
```

#### Update the implementation of the Batch GD to be Mini-Batch and rpeat the above plots and evaluations
#### Change batch size and see the effect

```python
def mini_batch_grad(inputs, targets, learn_rate, epoch, batch_size):
    losses = []
    thetas = []
    theta = np.atleast_2d(np.zeros((inputs.shape[1]))).T
    data = np.hstack((inputs, targets))
    batch_number = data.shape[0] // batch_size
    iter = 0
    for i in range(epoch):
        np.random.shuffle(data)
        for z in range(batch_number):
            iter += 1
            print("iteration : " + str(iter))
            mini_batch = data[z * batch_size:(z + 1)*batch_size, :]
            X_mini = data[:,0:-1]
            Y_mini = data[:,-1:]           
            outputs = np.dot(X_mini, theta)
            loss = np.mean((outputs - Y_mini)**2) / 2
            print("loss: " + str(loss))
            losses.append(loss)
            grads = np.atleast_2d(X_mini).T.dot(outputs-Y_mini) / len(Y_mini)
            theta = theta - learn_rate * grads
            thetas.append(theta)
            jac = np.linalg.norm(grads, 2)   
            print("jac: " + str(jac))    
            if jac < .0001 and jac > -.0001:
                break
    return np.array(thetas), losses
```

## Stochastic GD

#### Update the implementation of the Batch GD to beStochastic and rpeat the above plots and evaluations
#### You can try different learning rate and see its effect.
#### Compare to Batch and Mini-Batch GD

```python
def stochastic_grad(inputs, targets, learn_rate, epoch):
    losses = []
    thetas = []
    theta = np.atleast_2d(np.zeros((inputs.shape[1]))).T
    for i in range(epoch):
        print("epoch: " + str(i+1))
        for x, y in zip(inputs, targets):
            output = np.dot(x, theta)
            loss = ((output - y)**2) / 2
            print("loss: " + str(loss))
            losses.append(loss)
            #grads = 2 * np.atleast_2d(x).T.dot(output-y)
            grad0 = (output-y) * x[0]
            grad1 = (output-y) * x[1]
            grads = np.array([grad0, grad1])
            theta = theta - learn_rate * grads
            thetas.append(theta)
        jac = np.linalg.norm(grads, 2)   
        print("jac: " + str(jac))    
        if jac < .0001 and jac > -.0001:
            break 
    return np.array(thetas), losses
```
