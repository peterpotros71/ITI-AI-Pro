# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.linspace(0, 10, 100)
y = 3 + 1 * x  + np.random.normal(0, 1, size=100)

plt.figure(figsize=(10, 8));
plt.scatter(x, y, c='b', marker='x');
plt.show();
```

## Fit a line

Use the Moore-Penrose inverse to find the best line that fits the data.

```python
inputs = np.ones(x.shape[0]).reshape(-1,1)
inputs = np.append(inputs,x.reshape(-1,1),axis=1)
inputs.shape
```

```python
b, a = np.linalg.pinv(inputs) @ y
print(a)
print(b)
```

```python
plt.figure(figsize=(10, 8));
plt.scatter(x, y, c='b', marker='x');
plt.plot(x,a*x+b)
plt.show();
```
